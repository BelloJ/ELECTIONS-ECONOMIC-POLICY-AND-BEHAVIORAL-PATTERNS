Let's start by calculating the inflation rate for each election year.

```python
# Find the closest available dates to the election dates
closest_dates_to_election = {
    2011: all_items_index.index[all_items_index.index.searchsorted('2011-04-16')],
    2015: all_items_index.index[all_items_index.index.searchsorted('2015-03-28')],
    2019: all_items_index.index[all_items_index.index.searchsorted('2019-02-23')],
    2023: all_items_index.index[all_items_index.index.searchsorted('2023-02-25')],
}

# Calculate the inflation rate for each election year using the closest available dates
inflation_rate_election_years = {
    2011: (all_items_index.loc[closest_dates_to_election[2011]]['Monthly'] - all_items_index.loc[closest_dates_to_election[2011] - pd.Timedelta(days=365)]['Monthly']) / all_items_index.loc[closest_dates_to_election[2011] - pd.Timedelta(days=365)]['Monthly'] * 100,
    2015: (all_items_index.loc[closest_dates_to_election[2015]]['Monthly'] - all_items_index.loc[closest_dates_to_election[2015] - pd.Timedelta(days=365)]['Monthly']) / all_items_index.loc[closest_dates_to_election[2015] - pd.Timedelta(days=365)]['Monthly'] * 100,
    2019: (all_items_index.loc[closest_dates_to_election[2019]]['Monthly'] - all_items_index.loc[closest_dates_to_election[2019] - pd.Timedelta(days=365)]['Monthly']) / all_items_index.loc[closest_dates_to_election[2019] - pd.Timedelta(days=365)]['Monthly'] * 100,
    2023: (all_items_index.loc[closest_dates_to_election[2023]]['Monthly'] - all_items_index.loc[closest_dates_to_election[2023] - pd.Timedelta(days=365)]['Monthly']) / all_items_index.loc[closest_dates_to_election[2023] - pd.Timedelta(days=365)]['Monthly'] * 100,
}

inflation_rate_election_years
```

```python
# Convert the 'inflation_rate_election_years_new' dictionary to a dataframe
inflation_rate_df = pd.DataFrame(list(inflation_rate_election_years.items()), columns=['Year', 'Inflation Rate (%)'])

# Plot the inflation rate for each election year
plt.figure(figsize=(10, 6))

plt.plot(inflation_rate_df['Year'], inflation_rate_df['Inflation Rate (%)'], marker='o')

plt.title('Inflation Rate in Nigerian Presidential Election Years')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.grid(True)
plt.show()
```

The inflation rates for each election year using the closest available dates to the election dates are as follows:
